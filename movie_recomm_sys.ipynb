### Data-Processing

```python
# in the first step we are going to Join this two movie dataset
movies = movies.merge(credits,on='title')
movies.head(2)
```

```python
# let's check the dimensions of the dataset
movies.shape
```

```python
# Now, in This project we are going to build up a content based recommendation system, so we are choosing 
# those columns that are relevant to our project

selected_features = ['id','title','overview','genres','keywords','cast','crew']
movies_df = movies[selected_features] 
movies_df.head()
```

```python
# now we will check for presence of any null values inside the dataset
movies_df.isnull().sum()
```

```python
# there are only 3 null values present in the dataset, we are simply going to discard this 3 values
movies_df.dropna(inplace=True)
```

```python
# let's now check for any duplicated values in the dataset
movies_df.duplicated().sum()
# there are no duplicated values
```

```python
# now, our target is to extract the names of the genres and keywords from the given dataset. We are defining the following function to do this
import ast
def extract(obj):
    lst = []
    for dict in ast.literal_eval(obj):
        lst.append(dict['name'])
    return lst
```

```python
# applying the created function to the columns of 'genres' and 'keywords'
movies_df['genres'] = movies_df['genres'].apply(extract)
movies_df['keywords'] = movies_df['keywords'].apply(extract)
```

```python
movies_df.head()
```

```python
# in the next step we are going to get the top 3 casts of the movies, for this we are defining the following function
def get_top_3_casts(obj):
    lst = []
    count = 0
    for dict in ast.literal_eval(obj):
        if count<3:
            lst.append(dict['name'])
            count+=1
        else:
            break
    return lst
```

```python
# applying the created function to the column of 'cast'
movies_df['cast'] = movies_df['cast'].apply(get_top_3_casts)
```

```python
movies_df.head()
```

```python
# in the next step we are going to extract the director name from the following function
def get_director(obj):
    lst = []
    for dict in ast.literal_eval(obj):
        if dict['job'] == 'Director':
            lst.append(dict['name'])
            break
    return lst
```

```python
# applying the created function to the column of 'crew'
movies_df['crew'] = movies_df['crew'].apply(get_director)
```

```python
movies_df.head()
```

```python
# now we are going to convert the overview column into a list
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split(' '))
```

```python
movies_df.head()
```

```python
# we are Noe going to remove the spaces between the elements in the columns 'genres', 'keywords', 'cast', 'crew'
movies_df['genres'] = movies_df['genres'].apply(lambda x: [element.replace(' ','') for element in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [element.replace(' ','') for element in x]) 
movies_df['cast'] = movies_df['cast'].apply(lambda x: [element.replace(' ','') for element in x]) 
movies_df['crew'] = movies_df['crew'].apply(lambda x: [element.replace(' ','') for element in x]) 
```

```python
movies_df.head()
```

```python
# let's create a new column 'tags' that contains the concatenation of all the columns except id and title
movies_df['tags'] = movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
movies_df.head()
```

```python
# let's create a new df containing  'id', 'title' and 'tags'
movies_df_new = movies_df[['id','title','tags']]
movies_df_new.head()
```

```python
# let's join the tags into a string
movies_df_new['tags'] = movies_df_new['tags'].apply(lambda x:' '.join(x))
movies_df_new.head()
```

```python
# we are going to convert the tags into lower cases
movies_df_new['tags'] = movies_df_new['tags'].apply(lambda x:x.lower())
movies_df_new.head()
```

```python
# now we are going to use the stemming techniques so that in tags paragraph all the similar words like 'dance' dancing' 'dancer'
# will be converted  in to a root word. We are applying this by the use of the following function

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def stem(para):
    stemmed = []
    for words in para.split():
        stemmed.append(stemmer.stem(words))
    return ' '.join(stemmed)
```

```python
# stemming the 'tags'
movies_df_new['tags'] = movies_df_new['tags'].apply(stem)
movies_df_new.head()
```

```python
# now we are going to vectorize the tags using bagging techniques
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000, stop_words='english')
vectors = vectorizer.fit_transform(movies_df_new['tags']).toarray()
vectors
```

```python
# in the Next step we are going to get the similarity of one  movie with all the other movie using the cosine similarity, applied
# on the vectors
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python
# now we are going to create a function that will recommend top 5 movies after inserting a movie name
def recommend(movie):
    index = movies_df[movies_df['title'] == movie].index[0]
    top_5_recomm = sorted(list(enumerate(similarity[index])), key=lambda x:x[1], reverse=True)[1:6]
    movies_name = [movies_df_new.iloc[movie[0],]['title'] for movie in top_5_recomm]
    for recom in movies_name:
        print(recom)
```

```python
recommend('Hulk')
```
